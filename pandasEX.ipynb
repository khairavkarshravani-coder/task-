# Combining Dataframes - Inner Merge
- The .merge() method takes in a key argument labelled how
- There are 3 main ways of merging tables together using the how parameter
- 
    - Inner
    - 
     -Row is included only if the key exists in BOTH tables
    - Unmatched rows are removed
    -“Common data only”

    - Outer:-
  
    - “Everything from both tables”
    - Keeps all rows from both DataFrames
   
    - Left or Right

     - “Left table full”,Right table full”

```python
# registrations = pd.DataFrame({'reg_id': [1, 2, 3, 4], 'name': ['Andrew', 'Bobo', 'Claire', 'David']})
# logins = pd.DataFrame({'log_id': [1, 2, 3, 4], 'name': ['Xavier', 'Andrew', 'Yolanda', 'Bobo']})

registeration=pd.DataFrame({'reg_id':[1,2,3,4],'name':['shravani','prachi','uday','sujata']})
logins=pd.DataFrame({'login_id':[1,2,3,4],'name':['xoer','bebo','uday','shravani']})
```

```python
registeration
```

```python

logins
```

```python
pd.merge(registeration,logins,how='inner',on='name')
```

# Combining the DataFrame -left and right

```python
pd.merge(left=registeration,right=logins,how='left',on='name')
```

```python
pd.merge(left=registeration,right=logins,how='right',on='name')
```

# Comebining Dataframe -Outer merge

```python
pd.merge(left=registeration,right=logins,how='outer',on='name')
```

```python
pd.merge(right=registeration,left=logins,how='outer',on='name')
```

```python
pd.merge(registeration,logins,how='outer',on='name')
```

```python
# registrations = registrations.set_index('name')
registeration = registeration.set_index('name')
```

```python
registeration
```

```python
pd.merge(registeration , logins , how='outer' , left_index=True , right_on='name')
```

```python
registeration.columns=['reg_name','reg_id']
registeration
```

```python
pd.merge(registeration,logins,how='outer',left_on='reg_name',right_on='name')
```

# Pivot tabel

A pivot tabel is used to summerize data by ,convert long data ->summery tabel.
pivot table is used to summarize and analyze data by grouping rows and columns using aggregation functions.   

```python
df = pd.read_csv('Files\Licenses.csv')
```

```python
df.head()
```

```python
licenses = df[['Company','Product','Licenses']]
```

```python
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

```python
pd.pivot_table(df,index='Company',aggfunc='sum') #aggregation function(mean,max,mode,max)
```

```python
pd.pivot_table(df,index='Company',aggfunc='sum')[['Licenses','Sale Price']]
```

```python
df.groupby(['Company'])[['Licenses','Sale Price']].sum()
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='sum',columns='Product',fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='mean',columns='Product',fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='max',columns='Product',fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum,np.mean],fill_value=0)
```
