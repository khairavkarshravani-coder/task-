```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var =[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1=pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)
    
plt.plot(var,var_1)
plt.show()

       
```

# Bar plot Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python

sns.set(style="darkgrid")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order = order_1,hue_order=["Female","Male"],
           saturation=20,errwidth=0.3,alpha=0.5)
            

plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,color="g",
           palette="icefire",saturation=0.5)

plt.title("Vertical bar")
plt.show()
```

# Histogram plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230],
            kde=True,rug=True,color="g")
plt.show()

# bins=catgries wise histogram
# kde=line 
# rug= down line


```

#  Scatter plot

```python
import matplotlib.pyplot as plot
import seaborn as sns 
import numpy as np
import pandas  as pd
```

```python
var=sns.load_dataset ("penguins")
var
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm",data=var,hue="sex",
                style="sex",size="sex",sizes=(40,50),
               palette='gist_ncar',alpha=1,markers=m)
plt.show()
```
