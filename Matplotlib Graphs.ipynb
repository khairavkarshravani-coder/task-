# Stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,3,4,7,8,4]

plt.stem(x,y,linefmt=":",markerfmt="r" , bottom=0,basefmt="g",label="python")                  #(linefmt)- dotted line 
plt.legend()  #label show                                                                      # (markerfmt-r )- red plot
plt.show()                                                                                     #(Bottom=2)- satring plot 2 se start
                                                                                               #(basefmt)- base line coluer change  
                                                                                               #  orientation="horizontal" 
```

# step Plot

```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y= [8,6,3,9,0]

plt.step(x,y,color="r",marker="o",ms=10 ,mfc="g",label="Python")  # ms(marker size),mfc(marker face )
plt.grid()  
plt.xlabel("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2) # title print 
plt.grid()
# fig = plt.figure(figsize=(5,5)) 
plt.show()
```

# pie Chart

```python
import matplotlib.pyplot as plt
x=[0,3,4,6,8]
y=[6,0,4,7,8,9]
plt.plot(x)

plt.title("WScube",fontsize=10)
# plt.xlabel("days",fontsize=10)
# plt.ylabel("Python",fontsize=10)

plt.text(2,5,"python",fontsize=10,style="italic",bbox={"facecolor":"green"})
plt.annotate("python",xy=(3,6),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend(["up"],loc=9,facecolor="red",edgecolor="yellow",framealpha=0.3,)#shadow=True)

plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.title("Wscube")
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)

# plt.bar(x,y,width=0.3,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.4)

plt.bar(x,y,width=0.3,color="r",edgecolor='r',label="popularity")
plt.bar(x,z,width=0.3,color="y",edgecolor='r',label="popularity")
plt.legend()    # label under graph 
c=["y","b","m","g"]

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("Wscube")
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)


plt.bar(p,y,width,color="r",edgecolor='r',label="popularity")
plt.bar(p1,z,width,color="y",edgecolor='r',label="popularity")


plt.legend()    # label under graph 
plt.show()
```
